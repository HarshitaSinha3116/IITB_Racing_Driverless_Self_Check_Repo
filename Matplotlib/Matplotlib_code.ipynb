```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

gas = pd.read_csv('gas_prices.csv')

plt.plot(gas.Year, gas.USA, 'r-.', label='USA')
plt.plot(gas.Year, gas.Canada, 'b-.', label='Canada')

plt.title('Gas prices in USA and Canada')
plt.xlabel('Year')
plt.ylabel('Dollars/Gallon')
plt.legend()

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
#countries = ['USA', 'Canada','France', 'Italy', 'Austria','Japan', 'Mexico']
for countries in gas:
    if( countries!='Year'):
        plt.plot(gas.Year, gas[countries], marker = '.', label = [countries])
plt.xlabel('Year')
plt.ylabel('Dollars/Gallons')

plt.savefig('Gas_price_fig.png', dpi = 300)

plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')#create a pie chart showing the % of players which are left legged and right legged
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels = ['left','right']
plt.pie([left, right], labels = labels, autopct='%.2f %%')
plt.show()
```
